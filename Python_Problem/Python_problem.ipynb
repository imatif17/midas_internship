<font size="5">Importing tweepy Library</font>

```python
import tweepy
```

<font size="5">Filling the credential obtained from twitter</font>

```python
consumer_key = "" 
consumer_secret = ""
access_key = ""
access_secret = ""
```

<font size="4">In the below box we declare a function get_all_tweets which fetch all the tweets(maximum of 3240) from the account user_id. 
user_id is the only parameter of the function which accepts the twitter user id from which data has to be fetched.
This function return the information fetched by the tweepy library.</font>

```python
def get_all_tweets(user_id = '@midasIIITD'):
    
    # Authorization to consumer key and consumer secret     
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    # Access to user's access key and access secret
    auth.set_access_token(access_key, access_secret)
    # Calling api
    api = tweepy.API(auth)

    #Initialize a list to hold all the Tweets
    alltweets = []  

    #Make initial request for most recent tweets (200 is the maximum allowed count)
    new_tweets = api.user_timeline(screen_name = user_id,count=200,tweet_mode = 'extended')

    #Saving the most recent tweets
    alltweets.extend(new_tweets)

    #Saving the id of the oldest tweet less one
    oldest = alltweets[-1].id - 1

    #Keep grabbing tweets until there are no tweets left to grab
    while len(new_tweets) > 0:
        print (f"getting tweets before {oldest}")

        #All subsiquent requests use the max_id param to prevent duplicates
        new_tweets = api.user_timeline(screen_name = user_id,count=200,max_id=oldest,tweet_mode = 'extended')

        #Save most recent tweets
        alltweets.extend(new_tweets)

        #Updating the id of the oldest tweet less one
        oldest = alltweets[-1].id - 1

        print (f"...{len(alltweets)} tweets downloaded so far")
    
    return alltweets
```

<font size="4">Calling the function get_all_tweets and storing the output of the function in a variable all_tweets.</font>

```python
all_tweets = get_all_tweets()
```

<font size="5">Initializing a list json_tweet to add all _json part of the tweets fetched</font>

```python
json_tweet = []
for i in range(len(all_tweets)):
    json_tweet.append(all_tweets[i]._json)
```

<font size="5">Dumping all the response to JSON file json_file</font>

```python
import json
json_file = json.dumps(json_tweet)
```

<font size="5">Saving a file to pc as output.jsoln</font>

```python
import jsonlines
with open('output.jsoln','w') as outfile:
        outfile.write(json_file)

outfile.close()
```

<font size="5">Opening the above saved file as j_file</font>

```python
j_file = jsonlines.open('output.jsoln')
```

<font size="5">Storing the required parameters to be shown in arrays</font>

```python
#Initializing different lists
text = []
date_and_time = []
number_of_likes = []
number_of_retweets = []
number_of_image = []
#For all the tweet responses
for item in j_file.read():
    text.append(item["full_text"])
    date_and_time.append(item["created_at"])
    number_of_likes.append(item["favorite_count"])
    number_of_retweets.append(item["retweet_count"])
    if ('media' in item["entities"]):
        number_of_image.append(1)
    else:
        number_of_image.append(0)
```

<font size="5">Creating a pandas dataframe to show the datas in tabular form</font>

```python
#Importing pandas library
import pandas as pd
#Creating dataframe
data = pd.DataFrame({"Text of Tweet":text,
              "Date and time of tweet":date_and_time,
              "Number of favorites/likes":number_of_likes,
              "Number of retweets":number_of_retweets,
              "Number of images in tweet":number_of_image})
```
